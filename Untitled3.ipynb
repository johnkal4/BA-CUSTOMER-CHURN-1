 Σκοπός  
Η παρούσα εργασία στοχεύει στην πρόβλεψη της απώλειας πελατών (customer churn) χρησιμοποιώντας ένα βασικό μοντέλο XGBoost χωρίς διαδικασία fine-tuning.

Βήματα Υλοποίησης  

1. Φόρτωση και Εξερεύνηση Δεδομένων
   - Φορτώνονται τα δεδομένα από το αρχείο.  
   - Εξετάζονται οι βασικές πληροφορίες και η κατανομή των χαρακτηριστικών.  

2. Προετοιμασία Δεδομένων
   - Διαχείριση κατηγορηματικών μεταβλητών με κατάλληλη κωδικοποίηση (encoding).

   - Διαχωρισμός των δεδομένων σε training και test σύνολα.  

3. Ορισμός και Εκπαίδευση Μοντέλου
   - Δημιουργείται το μοντέλο XGBoost Classifier με προκαθορισμένες παραμέτρους (`n_estimators=25`, `eval_metric='error'`).  
   - Εκπαιδεύεται το μοντέλο πάνω στα δεδομένα εκπαίδευσης.  

4. Αξιολόγηση Απόδοσης  
   - Υπολογίζεται η ακρίβεια (accuracy) στο test set.

- Το μοντέλο παρουσιάζει ικανοποιητική ακρίβεια (~80%) χωρίς βελτιστοποίηση των υπερπαραμέτρων.  
- Υπάρχουν περιθώρια βελτίωσης, κυρίως μέσω διαδικασιών fine-tuning που θα εξεταστούν ξεχωριστά.  
- Η ισορροπία μεταξύ precision και recall υποδεικνύει σχετικά καλή απόδοση στην πρόβλεψη και των δύο κλάσεων (churn/no churn).  
- Είναι σημαντικό να κατανοήσουμε τα δεδομένα και να προετοιμάσουμε σωστά τις μεταβλητές πριν την εκπαίδευση.

Η παρούσα υλοποίηση αφορά το βασικό μοντέλο χωρίς fine-tuning. Η διαδικασία fine-tuning θα παρουσιαστεί σε ξεχωριστό notebook για τη βελτίωση της απόδοσης.


```python
import pandas as pd
import xgboost as xgb
import numpy as np
```

```python
bookings = pd.read_csv('hotel_bookings_clean.csv')
```

```python
bookings.info()
```

```python
bookings.describe()
```

```python
bookings['is_canceled'].value_counts().plot(kind='bar')
```

```python
bookings['is_canceled'].value_counts()/bookings['is_canceled'].count()
```

```python
X, y = bookings.iloc[:,1:], bookings.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33, random_state=123)
```

```python
xgb_clf=xgb.XGBClassifier(random_state=123)
```

```python
xgb_clf.get_params()
```

```python
xgb_clf.set_params(n_estimators=10)
```

```python
xgb_clf.set_params(max_depth=3)
```

```python
xgb_clf.set_params(eval_metric='error')
```

```python
xgb_clf.fit(X_train, y_train)
```

```python
preds = xgb_clf.predict(X_test)
xgb_clf.get_params()
```

```python
accuracy = float(np.sum(preds==y_test))/y_test.shape[0]
```

```python
print("Baseline accuracy:", accuracy)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rs_param_grid = {
    'max_depth': list(range(3,12)),
    'alpha': [0, .001, .01, .1],
    'subsample': [0.25, 0.5, 0.75, 1],
    'learning_rate': np.linspace(0.01, 0.5, 10),
    'n_estimator': [10,25,40]
    }
```

```python
xgb_clf = xgb.XGBClassifier(eval_metric='error', random_state=123)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
xgb_rs = RandomizedSearchCV(estimator=xgb_clf, param_distributions=rs_param_grid, cv=3, n_iter=5, verbose=2)
```

```python
xgb_rs.fit(X_train, y_train)
```

```python
print("Best parameters found:", xgb_rs.best_params_)
print("Best accuracy found:", xgb_rs.best_score_)
```
